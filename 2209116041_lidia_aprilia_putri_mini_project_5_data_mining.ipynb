Visualisasi dataset World Happiness Report menggunakan peta dunia adalah salah satu cara efektif untuk menyajikan informasi tentang skor kebahagiaan di berbagai negara di seluruh dunia secara geografis.

Setiap negara diwakili oleh wilayah/region yang ditandai dengan warna yang bervariasi sesuai dengan skala warna untuk memperjelas skor kebahagiaan di setiap negara. Negara dengan warna yang lebih terang menunjukkan skor kebahagiaan yang lebih tinggi. Sementara warna yang lebih gelap menunjukkan skor kebahagiaan yang lebih rendah.

Sebagai contoh, skor kebahagiaan negara Canada sebesar 7.427. Oleh karena itu, wilayah negaranya berwarna cokelat terang. Berbeda dengan negara Syria yang berwarna ungu gelap bahkan mendekati warna biru tua karena memiliki skor kebahagiaan sebesar 3.006, jauh lebih rendah daripada negara Canada.

Dapat kita simpulkan bahwa negara Canada berada pada peringkat teratas dan negara Syria berada pada peringkat terbawah untuk tingkat kebahagiaan negaranya.

Melalui visualisasi ini diharapkan dapat membantu pengguna untuk mendapatkan wawasan terkait perbedaan kebahagiaan antar negara dan dapat digunakan sebagai dasar untuk analisis lebih lanjut tentang variabel yang memengaruhi kebahagiaan setiap negara.

````python
region_lists = list(df['Region'].unique())
region_happiness_ratio = []

for each in region_lists:
    region = df[df['Region'] == each]
    region_happiness_rate = region['Happiness Score'].sum() / len(region)
    region_happiness_ratio.append(region_happiness_rate)

data = pd.DataFrame({'region': region_lists, 'region_happiness_ratio': region_happiness_ratio})

sorted_region = data.sort_values(by='region_happiness_ratio', ascending=False)

fig = px.bar(sorted_region, y='region', x='region_happiness_ratio',
             labels={'region_happiness_ratio': 'Region Happiness Ratio'}, width=800,
             title='World Happiness Ratio Across Regions', color='region_happiness_ratio',
             color_continuous_scale='agsunset', orientation='h')
fig.update_yaxes(title='Region', tickangle=-45)
fig.update_xaxes(title='Region Happiness Ratio')

fig.show()
````

Barplot diatas terdiri dari batang horizontal, dimana tinggi setiap batang menunjukkan rasio kebahagiaan untuk masing-masing wilayah. Setiap batang diberi label yang menunjukkan nama wilayahnya.

Selain itu, setiap batang memiliki warna yang berbeda. Semakin terang warna batangnya, maka semakin tinggi rasio kebahagiaan wilayah tersebut. Sebaliknya, semakin gelap warna batangnya, maka semakin rendah rasio kebahagiaan wilayahnya.

Sebagai contoh, wilayah Australia dan New Zealand memiliki rasio diatas 7, maka batangnya berwarna kuning terang. Berbeda dengan batang wilayah Sub-Saharan Africa yang berwarna ungu gelap karena memiliki rasio di range 4-5, jauh lebih rendah daripada wilayah Australia dan New Zealand.

Dapat kita tarik kesimpulan bahwa warga yang tinggal di negara yang terletak pada wilayah Australia dan New Zealand jauh lebih bahagia daripada warga yang tinggal di negara yang terletak pada wilayah Sub-Saharan Africa.

Melalui visualisasi dataset World Happiness Report menggunakan barplot diharapkan dapat memberikan gambaran terkait rasio kebahagiaan di berbagai wilayah/region di seluruh dunia karena barplot memperlihatkan perbandingan langsung rasio kebahagiaan di berbagai wilayah.

````python
selected_columns = ['Happiness Score', 'Economy (GDP per Capita)',
                    'Family', 'Health (Life Expectancy)', 'Freedom',
                    'Trust (Government Corruption)', 'Generosity']

df_selected = df[selected_columns]

fig = px.imshow(df_selected.corr(), color_continuous_scale='agsunset')
fig.update_layout(title="Correlation Heatmap of World Happiness Report Variables")

fig.show()
````

Setiap sel dalam heatmap mewakili korelasi antara dua variabel dalam dataset ini. Warna setiap sel menunjukkan tingkat korelasi dan arah korelasi antar variabel. Semakin terang warna selnya, maka semakin tinggi tingkat korelasinya. Sebaliknya, semakin gelap warna selnya, maka semakin rendah tingkat korelasinya.

Sebagai contoh, variabel Happiness Score dan variabel Economy (GDP per Capita) memiliki tingkat korelasi sebesar 0.7809655 dengan warna sel orange terang Artinya nilai dari kedua variabel ini sangat berpengaruh satu sama lain. Berbeda dengan variabel Happiness Score dan variabel Generosity yang memiliki tingkat korelasi sebesar 0.1803185 dengan warna sel ungu. Artinya nilai dari kedua variabel tidak cukup berpengaruh satu sama lain.

Melalui visualisasi dataset World Happiness Report menggunakan heatmap diharapkan dapat membantu dalam memahami korelasi antar variabel yang memengaruhi skor kebahagiaan setiap negara di seluruh dunia.

````python
region_lists = list(df['Region'].unique())
share_economy = []
share_family = []
share_health = []
share_freedom = []
share_trust = []
share_generosity = []

for region in region_lists:
    region_data = df[df['Region'] == region]
    share_economy.append(region_data['Economy (GDP per Capita)'].mean())
    share_family.append(region_data['Family'].mean())
    share_health.append(region_data['Health (Life Expectancy)'].mean())
    share_freedom.append(region_data['Freedom'].mean())
    share_trust.append(region_data['Trust (Government Corruption)'].mean())
    share_generosity.append(region_data['Generosity'].mean())

df_bar = pd.DataFrame({
    'Region': region_lists,
    'Economy': share_economy,
    'Family': share_family,
    'Health': share_health,
    'Freedom': share_freedom,
    'Trust': share_trust,
    'Generosity': share_generosity
})

fig = px.bar(df_bar, y='Region', x=['Economy', 'Family', 'Health', 'Freedom', 'Trust', 'Generosity'],
             barmode='stack',
             orientation='h',
             title='Factors Affecting Happiness Score Across Regions',
             labels={'value': 'Percentage of Region', 'variable': 'Factors'},
             color_discrete_sequence=px.colors.qualitative.Plotly)

fig.show()
````

Dalam stacked barplot, setiap batang mewakili satu region/wilayah, sedangkan bagian-bagian dari setiap batang mewakili besaran nilai dari variabel-variabel yang ada pada dataset World Happiness Report, seperti economy, family, health, freedom, trust, dan generosity. Bagian yang lebih tinggi menunjukkan bahwa variabel tersebut berpengaruh besar terhadap tingkat kebahagiaan suatu wilayah/region. Selain itu, setiap variabel diwakili oleh warna yang berbeda.

Sebagai contoh, pada wilayah Southern Asia, variabel yang berpengaruh besar terhadap tingkat kebahagiaan wilayah ini adalah variabel Family, dan variabel yang berpengaruh kecil yaitu Trust.

Dapat kita tarik kesimpulan bahwa faktor dukungan sosial dari keluarga, kerabat, dll merupakan poin terpenting dalam meningkatkan kebahagiaan para warga yang tinggal di negara yang terletak pada wilayah Southern Asia.

Melalui visualisasi stacked barplot, diharapkan dapat memudahkan pengguna untuk memahami dan membandingkan besaran nilai masing-masing variabel terhadap kebahagiaan suatu wilayah/region.

#####**Validasi Silang**

````python
inertia_values = []

k_range = range(2, 10)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(x_final_norm)
    inertia_values.append(kmeans.inertia_)

plt.plot(k_range, inertia_values, marker='o')
plt.xlabel('Number of clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.xticks(k_range)
plt.show()
````

Output diatas merupakan implementasi metode siku (elbow method) untuk menentukan jumlah optimal dari klaster dalam algoritma KMeans. Tujuan dari metode siku adalah untuk menemukan titik "siku" di plot, yaitu titik dimana penurunan nilai inertia tidak lagi signifikan saat jumlah klaster ditambahkan. Titik siku ini menunjukkan jumlah klaster optimal.


Berdasarkan hasil plot diatas, dapat kita lihat jika titik "siku" berada di angka 6. Dapat kita simpulkan jika jumlah klaster optimal dalam algoritma KMeans adalah 6.

####**ROC Curve**

````python
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelBinarizer

# Membuat objek LabelBinarizer
lb = LabelBinarizer()

# Mengonversi label menjadi bentuk one-hot encoding
y_test_bin = lb.fit_transform(y_test)

models = [knn, gnb, dtc]
model_names = ['K-Nearest Neighbor', 'Gaussian Naive Bayes', 'Decision Tree Classifier']

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for model, name, ax in zip(models, model_names, axes):
    y_pred_proba = model.predict_proba(x_test_norm)
    num_classes = y_pred_proba.shape[1]  # Jumlah kelas

    # Compute ROC curve and ROC area for each class
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    for i in range(num_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_pred_proba[:, i])
        roc_auc[i] = roc_auc_score(y_test_bin[:, i], y_pred_proba[:, i])

    # Compute micro-average ROC curve and ROC area
    fpr["micro"], tpr["micro"], _ = roc_curve(y_test_bin.ravel(), y_pred_proba.ravel())
    roc_auc["micro"] = roc_auc_score(y_test_bin, y_pred_proba, average="micro")

    ax.plot(fpr["micro"], tpr["micro"], label=f'{name} (micro-average ROC-AUC = {roc_auc["micro"]:.2f})', color='deeppink', linestyle=':')
    ax.plot([0, 1], [0, 1], color='navy', linestyle='--')
    ax.set_xlim([0.0, 1.0])
    ax.set_ylim([0.0, 1.05])
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_title(f'ROC Curve - {name}')
    ax.legend(loc='lower right')

plt.tight_layout()
plt.show()
````

Pada ROC Curve diatas, dapat kita simpulkan lagi bahwa model algoritma yang optimal dalam melakukan modelling predict/klasifikasi adalah algoritma K-Nearest Neighbor dengan nilai micro-average ROC-AUC sebesar 0.97.
